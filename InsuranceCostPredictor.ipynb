```python
import pandas as pd

insurance_dataset = pd.read_csv("C:/Users/MANSI/Desktop/DRF/Insurance/Model/healthinsurance.csv")
```

```python
insurance_dataset.replace({'sex':{'male':0 ,'female':1}},inplace=True)
insurance_dataset.replace({'smoker':{'yes':0,'no':1}},inplace=True)
insurance_dataset.replace({'region':{'southeast':0,'southwest':1,'northeast':2,'northwest':3}},inplace=True)
```

```python
X = insurance_dataset.drop(columns='charges',axis=1)
Y = insurance_dataset['charges']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y ,test_size=0.2 , random_state=2)
```

```python
print(X.shape ,X_train.shape , X_test.shape)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, Y_train)
```

```python
training_data_perdiction = regressor.predict(X_train)
```

```python
r2_train = metrics.r2_score(Y_train,training_data_perdiction)
print('R sqaured value: ',r2_train)
```

```python
test_data_prediction = regressor.predict(X_test)
```

```python
r2_test = metrics.r2_score(Y_test,test_data_prediction)
print('R sqaured value: ',r2_test)
```
